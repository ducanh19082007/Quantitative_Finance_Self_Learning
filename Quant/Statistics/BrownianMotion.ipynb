# The ideas behind the code

## 1. Brownian Motion

the pure definition of Brownian motion from what i have learned is $$dW = \sqrt(dt) \phi$$ as phi being the normal distribution randomized values

hence, we can conclude that $$W_t = W_0 + \sqrt(dt) \phi$$

in which $W_0$ is equal to 0 usually, but for some cases like lognormal model, it should not be 0

```python
def BrownianMotion():
    n = 10000 #steps
    T = 1 #the final times
    times = np.linspace(0., T, n) #timestep arrays a evenly distributed range between 0 to T
    dt = times[1] - times[0] # timesteps, given that t1 - t0 is the VARIANCE of the normal distribution
    dW = np.random.normal(loc = 0.0, scale = np.sqrt(dt), size = (n-1,)) #normal simulation phi the sizes represents dimension, or this is gonna be sqrt dt * phi
    #example of the np.random.normal(size = (n-1,)) [a,b,c,d,e,..] each values in the array is normal distributed
    W0 = np.zeros(shape=(1,)) #given that initially B0 should be all 0, this could be changed based on the model one chose
    W_t = np.concatenate((W0, np.cumsum(dW))) 
    plt.plot(times, W_t)
    plt.show()
    
BrownianMotion()
    
    
    
```

# The Quadratic Variation

$${\displaystyle [X]_{t}=\lim _{\Vert P\Vert \rightarrow 0}\sum _{k=1}^{n}(X_{t_{k}}-X_{t_{k-1}})^{2}}$$

the scary P value min over there is just in case one have a certain distruptions and hence uwill be false, and when the occurs just generate new difference with Euler-Mảuyama step

```python
import numpy as np
import matplotlib.pyplot as plt

def BrownianMotion_QuadVarSamples():
    n = 10000
    d = 10
    T = 1 
    times = np.linspace(0., T, n)
    dt = times[1] - times[0]
    dW = np.random.normal(loc = 0.0, scale = np.sqrt(dt), size = (n-1, d))
    W0 = np.zeros(shape=(1, d))
    W_t = np.concatenate((W0, np.cumsum(dW, axis = 0)), axis = 0) 
    QV_W = np.cumsum(np.power(np.diff(W_t, axis = 0, prepend = W0), 2), axis = 0)
    
    plt.plot(times, QV_W)
    plt.plot(times, W_t)
    plt.show()
    
BrownianMotion_QuadVarSamples()
```
