```python
import numpy as np
import math
import pandas as pd
import datetime as dt
import scipy.stats as stats
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
stock_list = ['CBA.AX', 'BHP.AX', 'TLS.AX', 'NAB.AX', 'WBC.AX', 'STO.AX'] #pick some random Australian stocks
endDate = dt.datetime.now()
startDate = endDate - dt.timedelta(days = 300)

def get_data(stock_list, start, end):
    stock_data = {}
    for stock in stock_list:
        ticker = yf.Ticker(stock)
        stock_data[stock] = ticker.history(start=start, end=end)['Close']
    
    stock_data = pd.DataFrame(stock_data)
    log_return = np.log(stock_data / stock_data.shift(1))
    log_return = log_return[1:]
        
    meanReturn = log_return.mean()
    covariance_matrix = log_return.cov()
    
    return meanReturn, covariance_matrix, stock_data

mean_return, covariance_matrix, stock_data = get_data(stock_list, startDate, endDate)
print("Mean return: ", mean_return, "\n\n\n covariance matrix \n", covariance_matrix, "\n\n\n stock data by 300 days \n", stock_data) 
```
