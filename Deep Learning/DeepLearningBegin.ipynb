###Leaky ReLU

So why while we have normal Relu, we have something called Leaky ReLU. This is because when using ReLU, we can risk having dead neurons, when all neurons = 0, thus it could stop learning.

normal or Relu in general, does not treats negative and positive values quite fairly. Thus increasing the risk of inefficiency and slowing down the learning process.

<H3>Introduction to Leaky ReLU</h3>

Leaky ReLU is a modified version of ReLU designed to fix the problem of dead neurons. So Instead of returning zero for negative imputs. It allow a small fixed slope to the negative part of the input, ensures that neurons do not become inactive during training and though the changes in the weight between those small neurons could be insigificant, it still meant that the model will kept on learning. instead of giving up.

So this is the principal of leaky ReLU

$$
LeakyReLU =
\left\{
\begin{aligned}
x, (if \: x≥0)\\
negative\:slope × x \: \: \:(otherwise)
\end{aligned}
\right.
$$

##Relations to Gradient Descent

<h2> Understanding the Gradient ∇C in Backpropagation</h2>

$$
\nabla C =
\begin{bmatrix}
\frac{\partial C_o}{\partial w^{(1)}} \\
\frac{\partial C_o}{\partial b^{(1)}} \\
\vdots \\
\frac{\partial C_o}{\partial w^{(L)}} \\
\frac{\partial C_o}{\partial b^{(L)}}
\end{bmatrix}
$$

This is the <strong>gradient vector</strong> of the cost function $C$, where:
- $w^{(l)}$: weights of layer $l$
- $b^{(l)}$: biases of layer $l$
- $L$: the total number of layers in the neural network

---

<h3> How does all of these equations relate to gradient descent?</h3>

Gradient descent uses these gradients to <strong>update the weights and biases</strong> to reduce the cost function:

$$
w^{(l)} := w^{(l)} - \eta \cdot \frac{\partial C_o}{\partial w^{(l)}}
$$

$$
b^{(l)} := b^{(l)} - \eta \cdot \frac{\partial C_o}{\partial b^{(l)}}
$$

Where $\eta$ is the <strong>learning rate</strong>.

So, the vector $\nabla C$ is like a <strong>map of directions</strong>: it tells you how to move every single parameter (weights and biases across all layers) in order to <strong>descend</strong> the loss surface.
